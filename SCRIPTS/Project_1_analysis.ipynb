#VADER Sentiment Analysis

```python
# installing analysis package
# pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

#Simone Biles

```python
# Create an instance of the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Define a function to perform sentiment analysis using VADER
def get_sentiment(tokens):
    sentiment = analyzer.polarity_scores(tokens)
    compound_score = sentiment['compound']
    return compound_score

# Apply the function to the 'tokens' column of the DataFrame
clean_sb['sentiment'] = clean_sb['Comment Text'].apply(get_sentiment)

# Print the DataFrame with sentiment scores
print(clean_sb['sentiment'])
```

```python
# Count the occurrences of each sentiment
sentiment_counts = pd.cut(clean_sb['sentiment'], bins=3, labels=['Negative', 'Neutral', 'Positive']).value_counts()

# Plot the sentiments
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Analysis')
plt.show()

# Print the number of counts for each sentiment
for sentiment, count in sentiment_counts.items():
    print(f"{sentiment}: {count}")
```

#Playing with word clouds

```python
# import word cloud code
from wordcloud import WordCloud
```

Creating a Word Cloud for comments from Simone Biles' posts

```python
simone_cloud = " ".join(comment for comment in clean_sb['Comment Text'])

# Creating word_cloud with text as argument in .generate() method

simone_word_cloud = WordCloud(collocations = False, background_color = 'white').generate(simone_cloud)

# Display the generated Word Cloud

plt.imshow(simone_word_cloud, interpolation='bilinear')

plt.axis("off")

plt.show()
```

Creating a Word Cloud for comments under Jonathan Owens' posts

```python
jonathan_cloud = " ".join(comment for comment in clean_jo['Comment Text'])

# Creating word_cloud with text as argument in .generate() method

jonathan_word_cloud = WordCloud(collocations = False, background_color = 'white').generate(jonathan_cloud)

# Display the generated Word Cloud

plt.imshow(jonathan_word_cloud, interpolation='bilinear')

plt.axis("off")

plt.show()
```
